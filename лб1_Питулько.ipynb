2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями суми перших n натуральних чисел для кожного n, починаючи з 1, та виведіть його значення

```python
a = np.zeros((3, 3), dtype=int)

for i in range(3):
    for j in range(3):
        n = i * 3 + j + 1
        a[i, j] = n * (n + 1) // 2

print(a)
```

3.	Створіть масив розміром 7x7, заповніть його випадковими дійсними числами від 0 до 1 та знайдіть середнє геометричне його елементів.

```python
a = np.random.uniform(0, 2, (7,7))
geometric_mean = np.prod(a) ** (1 / a.size)

print("Матриця:")
print(a)
print("Середнє геометричне:", geometric_mean)
```


4.	Створіть масив розміром 6x6, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить мінімальні елементи в кожному рядку.



```python
a = np.random.rand(6, 6)

min = np.min(a, axis=1)
min_col = min.reshape(-1, 1)
print("Матриця:")
print(a)
print("Вектор-стовпчик з мінімальними елементами в кожному рядку:")
print(min_col)
```

5. 	Створіть два одновимірних масиви розміром 4, заповніть їх випадковими цілими числами від -5 до 5 та виконайте на них поелементні операції додавання, віднімання та множення.

6:.	Створіть два вектори розміром 4, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток.

8. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 12 та знайдіть її обернену матрицю

```python
a = np.random.randint(1, 13, size=(3, 3))
print("Вихідна матриця:")
print(a)

if np.linalg.det(a) == 0:
    print("Матриця не має оберненої матриці.")
else:
    inv_a = np.linalg.inv(a)
    print("Обернена матриця:")
    print(inv_a)
```

12.	Створіть дві матриці розміром 9x9, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть відношення елементів другого стовпця другої матриці до відповідних елементів першої матриці (поелементне ділення).

```python
a = np.random.randint(1, 11, (9, 9))
b = np.random.randint(1, 11, (9, 9))

print("Перша матриця:")
print(a[:, :3])
print("Друга матриця:")
print(b[:, :3])

result = b[:, 1] / a[:, 1]

print("Відношення елементів другого стовпця другої матриці до першої матриці:")
print(result)
```

14.	Створіть матрицю розміром 8x8, заповніть її випадковими цілими числами в діапазоні від 1 до 100 та знайдіть середнє геометричне значення елементів матриці.

```python
a = np.random.randint(1, 101, size=(8, 8))
print("Матриця:")
print(a)

geometric_avg = np.prod(a) ** (1 / a.size)

print("Середнє геометричне значення матриці", geometric_avg)
```

15.	Створіть матрицю розміром 10x10, заповніть її випадковими дійсними числами в діапазоні від -2 до 2 та знайдіть вектор-стовпчик, що містить найменше за модулем ціле значення в кожному рядку матриці.

```python
a = np.random.uniform(-2, 2, (10, 10))

print("Матриця:")
print(a)

min_abs_values = np.min(np.abs(a), axis=1)
min_index = np.argmin(np.abs(a), axis=1)

result = a[np.arange(10), min_index].reshape(-1, 1)

print("Вектор стовпчик з мінімальними за модулем ")
print(result)
```
